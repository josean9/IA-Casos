```python
v_continuas = []
v_categoricas = []
v_binarias = []
for i in df.columns:
    if df[i].nunique() == 2:
        v_binarias.append(i)
    elif df[i].nunique() > 55 and df[i].dtypes in ['float64', 'int64']:
        v_continuas.append(i)
    else:
        v_categoricas.append(i)

print('Variables continuas: {}'.format(', '.join(v_continuas)))
print('Variables categóricas: {}'.format(', '.join(v_categoricas)))
print('Variables binarias: {}'.format(', '.join(v_binarias)))
```

```python
df.drop(columns=["asma", "alcohol", "ayudas_publicas", "seguro_salud"], inplace=True)  # Eliminar la columna 'id' del DataFrame
```

```python
v_categoricas.remove("asma")
v_categoricas.remove("alcohol")
v_categoricas.remove("ayudas_publicas")
v_categoricas.remove("seguro_salud")
```

```python
# Hacer ordinal encoder en las variables categóricas actividad_fisica [Baja, Moderada, Alta], nivel_educativo [Primaria, Secundaria, Universitaria, Posgrado]
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[['Baja', 'Moderada', 'Alta'], ['Primaria', 'Secundaria', 'Universitaria', 'Posgrado']])
df[['actividad_fisica', 'nivel_educativo']] = encoder.fit_transform(df[['actividad_fisica', 'nivel_educativo']])
```

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        return 'Desconocido'  # Para países no reconocidos

df['continente_nacimiento'] = df['pais_nacimiento'].apply(country_to_continent)
v_categoricas.append('continente_nacimiento')
v_categoricas.remove("pais_nacimiento")
```

```python
# Columnas categóricas a codificar
columnas_categoricas = ['tipo_empleo', 'estado_civil', 'ocupacion', 'region', 'continente_nacimiento']

# Aplicar One-Hot Encoding con pandas
df = pd.get_dummies(df, columns=columnas_categoricas, drop_first=True)
v_categoricas.remove("tipo_empleo")
v_categoricas.remove("estado_civil")
v_categoricas.remove("ocupacion")
v_categoricas.remove("region")
v_categoricas.remove("continente_nacimiento")
```

```python
#hame un standard scaler en la variable gastos_salud en df["gastos_salud"]
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['gastos_salud'] = scaler.fit_transform(df[['gastos_salud']])

# Guardar el DataFrame modificado en un nuevo archivo CSV
```

```python
# en estas columnas quiero sustituir false por 0 y true por 1 tipo_empleo_Desempleado	tipo_empleo_Privado	tipo_empleo_Público	tipo_empleo_Temporal	estado_civil_Divorciado	estado_civil_Soltero	estado_civil_Viudo	ocupacion_Desempleado	ocupacion_Empleado	ocupacion_Jubilado	region_Este	region_Norte	region_Oeste	region_Sur	continente_nacimiento_Antarctica	continente_nacimiento_Asia	continente_nacimiento_Desconocido	continente_nacimiento_Europe	continente_nacimiento_North America	continente_nacimiento_Oceania	continente_nacimiento_South America
df.replace({'tipo_empleo_Desempleado': {False: 0, True: 1},
                'tipo_empleo_Privado': {False: 0, True: 1},
                'tipo_empleo_Público': {False: 0, True: 1},
                'tipo_empleo_Temporal': {False: 0, True: 1},
                'estado_civil_Divorciado': {False: 0, True: 1},
                'estado_civil_Soltero': {False: 0, True: 1},
                'estado_civil_Viudo': {False: 0, True: 1},
                'ocupacion_Desempleado': {False: 0, True: 1},
                'ocupacion_Empleado': {False: 0, True: 1},
                'ocupacion_Jubilado': {False: 0, True: 1},
                'region_Este': {False: 0, True: 1},
                'region_Norte': {False: 0, True: 1},
                'region_Oeste': {False: 0, True: 1},
                'region_Sur': {False: 0, True: 1},
                'continente_nacimiento_Antarctica': {False: 0, True: 1},
                'continente_nacimiento_Asia': {False: 0, True: 1},
                'continente_nacimiento_Desconocido': {False: 0, True: 1},
                'continente_nacimiento_Europe': {False: 0, True: 1},
                'continente_nacimiento_North America': {False: 0, True: 1},
                'continente_nacimiento_Oceania': {False: 0, True: 1},
                'continente_nacimiento_South America': {False: 0, True: 1}}, inplace=True)
```
